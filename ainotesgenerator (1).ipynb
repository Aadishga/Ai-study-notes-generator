```python
import google.generativeai as genai
from datetime import datetime

API_KEY = "YOUR_GEMINI_API_KEY_HERE"
genai.configure(api_key=API_KEY)
model = genai.GenerativeModel("gemini-3.5-flash")

def generate_study_notes(topic, level="beginner"):
    print(f"\n📚 Generating notes for: '{topic}' [{level}]...")
    prompt = f"""You are an expert educational content writer.
Create structured study notes on: {topic}
Level: {level}

Write these sections:
1. SUMMARY (3-4 sentences)
2. KEY CONCEPTS (5 bullet points)
3. IMPORTANT DEFINITIONS (4 terms)
4. REAL WORLD EXAMPLE (1 paragraph)
5. MINI QUIZ (3 questions with answers)
6. REMEMBER THIS (1 key takeaway)

Keep language simple for a {level} student."""

    response = model.generate_content(prompt)
    notes = response.text.strip()
    print(notes)
    
    filename = f"notes_{topic.replace(' ','_').lower()}.txt"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(notes)
    print(f"\n💾 Saved → {filename}")

generate_study_notes("einstein theory", "beginner")
```
